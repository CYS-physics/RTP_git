```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import trange

d=0.05

def v(x):
    #return 6.5*(x<0)-3.3*(x>=0)
    return 9.8/(1+np.exp(x/d)) -3.3

x = np.linspace(-50,50,1000)
plt.plot(x,v(x))
plt.scatter(-2,v(-2))
plt.scatter(-1,v(-1))
plt.scatter(0,v(0))
plt.scatter(1,v(1))
plt.scatter(2,v(2))
plt.show()
```

```python
dt = 0.01
x0 = -2
N=200
x_history = np.empty(N)
v_history = np.empty(N)
time = np.linspace(0,N*dt,N)

method = 'euler'

x_curr = x0

for i in trange(N):
    x_history[i] = x_curr
    v_history[i] = v(x_curr)
    
    if method=='euler':
        x_curr +=v(x_curr)*dt
    elif method=='RK4':
        k1 = v(x_curr)*dt
        x1 = x_curr+0.5*k1
        k2 = v(x1)*dt
        x2 = x_curr+0.5*k2
        k3 = v(x2)*dt
        x3 = x_curr+k3
        k4 = v(x3)*dt
        
        x_curr +=1/6*(k1+2*k2+2*k3+k4)
        

print(dt*6)
plt.scatter(time,x_history,s=10)
plt.plot(time,x_history, label='x')
plt.legend()
plt.show()
plt.plot(time,v_history, label='v')
plt.legend()
plt.show()
```
